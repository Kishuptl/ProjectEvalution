# Use of VIF for multicollinearity

```python
import statsmodels.api as sm
from scipy import stats

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif_cal(x1):
    vif = pd.DataFrame()
    vif['Variables'] = x1.columns
    vif['VIF factor'] = [variance_inflation_factor(x1.values , i) for i in range(x1.shape[1])]
    return(vif)

vif_cal(x1)
```

As per above we can see that year and total bags has highest multicollinearity present , so we need to drop that columns.

```python
x1 = avdf.drop(['year','Total Bags'] , axis=1)
vif_cal(x1)
```

now we can see that multicollinearity is removed .
